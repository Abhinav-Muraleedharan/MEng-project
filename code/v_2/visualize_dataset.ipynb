```python
## Load dataset
dataset = NWBDataset("dataset/sub-Indy", "*train", split_heldout=False)
```

```python
## View 'dataset.data'
dataset.data
```

```python
dataset.data.iloc[:,15]
```

```python
dataset.data["spikes"]
```

```python
df_neural_input = dataset.data["spikes"]
```

```python
# Identifying the NaN locations
nan_locations = df_neural_input.isna()

# Printing the locations of NaN values
for column in nan_locations:
    rows_with_nan = nan_locations[column][nan_locations[column]].index.tolist()
    for row in rows_with_nan:
        print(f"NaN found at Row {row}, Column '{column}'")
```

```python
X_neural = df_neural_input.to_numpy()
```

```python
print(X_neural)
```

```python
contains_nan = np.any(np.isnan(X_neural))
```

```python
print(contains_nan)
```

```python
row_sum_vector = np.sum(X_neural, axis=1)

# Perform column sum
column_sum_vector = np.sum(X_neural, axis=0)

# Display the vectors
print("Row Sum Vector:")
print(row_sum_vector.shape)
print(column_sum_vector.shape)
```
